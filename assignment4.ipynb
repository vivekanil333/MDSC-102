```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df=pd.read_csv("wine.data")
column_names=["class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium"," Total phenols"," Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","	Hue","OD280/OD315 of diluted wines","Proline "]
df=pd.read_csv("wine.data",names=column_names)
df
```

```python
sns.scatterplot(x="Alcohol",y="Ash",data=df,hue="class")
```

```python
sns.countplot(x = df['Alcalinity of ash'])
plot.show()

# Count plot for the feature 'class' of the alcohols
sns.countplot(data = df, x = 'class')
plot.title("Count plot of the feature 'Class' of alcohol")
```

```python
fig, axes = plot.subplots(4,3, figsize = (30,40)) 
i = 1
for feature in df:
  if feature != 'Alcohol' and feature != 'class':
    plot.subplot(4,3,i)
    sns.scatterplot(x = df[feature], y = df['Alcohol'], hue = df['class'])
    plot.grid()
    i = i+1
```
